```python
def execute_query_and_print(cursor, query):
        """
        Executes a query and prints results
        :param cur: the Cursor object
        :param query: the sql query to be executed
        :return:
        """
        cursor.execute(query)
     
        rows = cursor.fetchall()
     
        for row in rows:
            print(row)
```

```python
execute_query_and_print(cursor,"select user from election2")
```

```python
execute_query_and_print(cursor,"SELECT text FROM twitter.retweeted_status group by (id)")
```

```python
execute_query_and_print(cursor,"select name from twitter.usersinfo order by followers_count")
```
